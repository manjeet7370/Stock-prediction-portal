```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Type of Plotes

#### Line plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot')
plt.xlabel('X-axis lable')
plt.ylabel('Y-axis lable')
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize = (10,6))
plt.bar(months, revenue)
plt.title('Monthly Revenue Bar Chart')
plt.xlabel('Months')
plt.ylabel('Revenue')
```

## Histogram

```python
marks50 = np.random.randint(0, 100, 50)
marks50
```

```python
plt.hist(marks50);
plt.xlabel('Marks')
plt.ylabel('No of Students')
```

```python
plt.hist(marks50, bins=10)
```

### Scatter Plote

```python
height = [65, 70, 68, 72, 66]
weight = [150, 160, 155, 180, 145]
```

```python
plt.scatter(height, weight)
plt.title('Height vs Weight scatter Plot')
plt.xlabel('height (inches)')
plt.ylabel('weight (pounds)')
```

## Subplots and legends

```python
days = [1,2,3,4,5]
orignal_price = [10, 20 ,30, 40, 45]
prediction_price = [8,21,25,28,38]
```

```python
plt.plot(days, orignal_price, 'b-o', label='Orignal Price')
plt.plot(days, prediction_price, 'r-o', label='Prediction price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('weekends')
plt.ylabel('Price')
```
